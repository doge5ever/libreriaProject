```python
import numpy as np
import pandas as pd
import re
import time
import unidecode

df = pd.read_csv('./books_to_scrape/spiders/scraped_data.csv')
df.head(15)
```

```python
def process_product_desc(string):
    if string[-7:] == '...more':
        string = string[:-7]
    return unidecode.unidecode(string)
    
def process_availability(string):
    string = re.findall("\d", string)
    return int("".join(string))

def process_rating(string):
    mapping = {"One": 1, "Two": 2, "Three": 3, "Four": 4, "Five": 5}
    return mapping[string]

def process_price(string):
    string = string[1:]
    price = round(1.25 * float(string), 2)
    return price

df.drop(['product_type', 'tax', 'number_of_reviews', 'images'], axis=1, inplace=True)

df.dropna(axis=0, how='all', inplace=True)
df.rating = df.rating.apply(process_rating)
df.product_desc = df.product_desc.apply(process_product_desc)

# (df.price_excl_tax == df.price_incl_tax).all()
df.drop('price_excl_tax', axis=1, inplace=True)
df.rename(columns={"price_incl_tax": "price_USD"}, inplace=True)

df.price_USD = df.price_USD.apply(process_price)
df.availability = df.availability.apply(process_availability)
df.reset_index(drop=True, inplace=True)

df.head(15)
```

```python
df.to_json('dataset.json', orient="records")
df[:14].to_json('sample_dataset.json', orient="records")
```
